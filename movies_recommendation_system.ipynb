#Natural Language Processing

```python
df['tags'][0]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #remove punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)
  #tokenisation
  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]
  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

TF-IDF Vectorization

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index, index = df['title']).drop_duplicates()
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1,2), stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

Cosine Similarity

```python
# virat kohli
# virat
# vobulary of these 2 = virat, kohli  1) 1,1   2) 1,0
# we will plot these pts and determine the angle
# cos of that angle(cos(45)) = (0.71) cosine similarity score of those 2 sentences
# 1) we used graph bcs statements were small but in real statements will be big
# for that we have formula
# cs = ( summation(Ai * Bi)] ) / ( [root(summtion(Ai^2))] * [root(summation(Bi^2))] )
#  all summation i from 1 to n
#
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n=10):
  if title not in indices:
    return ['Movie not found']
  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```
